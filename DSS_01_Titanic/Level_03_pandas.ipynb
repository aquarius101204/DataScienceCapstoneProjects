```python
file_url = "https://bit.ly/ds01-weight-history"
data = pd.read_csv(file_url, index_col = "회차")
```

## 기본연산

```python
data["몸무게"].mean()
data["몸무게"].min()
data["몸무게"].max()
data["몸무게"].describe()
```

```python
data["지점"].replace("강남구", "강남")
```

```python
data
```

```python
data["지점"] = data["지점"].replace("강남구", "강남")
```

```python
data
```

```python
data["몸무게"] = data["몸무게"].fillna(75.8)
```

```python
data
```

```python
data["몸무게(prev)"] = data["몸무게"].shift(+1)
data["몸무게(next)"] = data["몸무게"].shift(-1)
data
```

```python
data["측정일"] = pd.to_datetime(data["측정일"])
data["측정일"].dt.year
```

```python
data["측정일"].dt.month
```

```python
data["측정일"].dt.day
```

## 컬럼 추가 & 수정

```python
data["지역"] = "서울"
data
data["지역"] = "부산"   ## 수정은 추가와 완전 동일하다.
data
```

```python
data["결제"] = [True, True, False, True, False, False, True, True, True]
data
```

```python
data["몸무게(gram)"] = data["몸무게"] * 1000
data
```

```python
data["과체중"] = data["몸무게"] > 75.8
data
```

```python
data.loc[data["몸무게"] > 75.8, "상태"] = "과체중"
data.loc[data["몸무게"] <= 75.8, "상태"] = "정상"

data
```
