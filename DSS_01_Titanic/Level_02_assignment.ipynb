```python
gallon = 75

if gallon <= 20:
    price = 90 * gallon
elif 20 < gallon <= 60:
    price = 90 * 20 + 80 * (gallon - 20)
else:
    price = 90 * 20 + 80 * 40 + 75 * (gallon - 60)

price = str(price / 100)
print(gallon, "$"+price)
```

```python
goal = 500
clap = "Clap! "

for n in range(goal):
    n_text = str(n)
    
    if "3" in n_text or "6" in n_text or "9" in n_text:
        repeat = n_text.count("3") + n_text.count("6") + n_text.count("9")
        for i in range(repeat):
            clap = repeat * "Clap! "
            print(n, clap)
    else:
        print (n)
```
